```python
from selenium import webdriver
import time
import pandas as pd
import re
import ast
```

```python
def clean_num(num_text):
    num_text = num_text.replace(',', '')
    multiplier = 1
    if 'K' in num_text:
        multiplier = 1000
    elif 'M' in num_text:
        multiplier = 1000000
    try:  
        num = ast.literal_eval(re.findall(r'\d+(?:\.\d+)?', num_text)[0]) * multiplier
    except:
        num = num_text
    return num
```

```python
clean_num("+1.2K")
```

```python
# Set up the path to the chrome driver
PATH = "PATH_TO_CHROME_DRIVER"
driver = webdriver.Chrome(PATH)
a = int(input("Enter starting page: "))
b = int(input("Enter Ending page: "))
all_details = []
for page_no in range(a, b + 1):
    try:
        # Get current page
        driver.get("https://twitchtracker.com/channels/ranking/french?page={}".format(page_no))
        print("\nPage:\nhttps://twitchtracker.com/channels/ranking/french?page={}".format(page_no))
        records = driver.find_elements_by_tag_name('tr')
        
        for record in records:
            
            fields = record.find_elements_by_tag_name('td')
            if len(fields) < 11:
                continue
            
            extracted_record = {
                'name': '',
                'avg_viewers': '',
                'time_streamed': '',
                'all_time_peak_viewers': '',
                'hours_watched': '',
                'rank': '', 
                'follower_gained': '',
                'total_followers': '',
                'total_views': ''
                
            }
            extracted_record['name'] = fields[2].text
            extracted_record['avg_viewers'] = clean_num(fields[3].text)
            extracted_record['time_streamed'] = clean_num(fields[4].text)
            extracted_record['all_time_peak_viewers'] = clean_num(fields[5].text)
            extracted_record['hours_watched'] = clean_num(fields[6].text)
            extracted_record['rank'] = clean_num(fields[7].text)
            extracted_record['follower_gained'] = clean_num(fields[8].text)
            extracted_record['total_followers'] = clean_num(fields[9].text)
            extracted_record['total_views'] = clean_num(fields[10].text)
            
            # Append record to all details
            all_details.append(extracted_record)
            

    except Exception as e:
        # Close the browser if we run to an error
        print(e)
        driver.close()
```

```python
df = pd.DataFrame(all_details)
df
```

```python
# Save the information into a CSV file
df.to_csv("Ranking_Table.csv", sep=',')
```
