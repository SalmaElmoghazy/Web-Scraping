```python
from selenium import webdriver
import time
import pandas as pd
```

```python
# Set up the path to the chrome driver
PATH = "E:\chrome_driver_32bit\chromedriver.exe"
driver = webdriver.Chrome(PATH)
all_details = []

driver.get("https://lcgpa.gov.sa/ar/Regulations/Docs-Lists/Pages/FactoriesList.aspx")

while True:
    try:
        
        records = driver.find_elements_by_tag_name('tr')
        print(len(records))
        
        records = records[1: ]
        for record in records:

            fields = record.find_elements_by_tag_name('td')

            extracted_record = {
                'رمز الفئة': '',
                'اسم المصنع': '',
                'العنوان': '',
                'رقم التواصل': '',
                'البريد الالكتروني': '',

            }
            extracted_record['رمز الفئة'] = fields[0].get_attribute('textContent')
            extracted_record['اسم المصنع'] = fields[1].get_attribute('textContent')
            extracted_record['العنوان'] = fields[2].get_attribute('textContent')
            extracted_record['رقم التواصل'] = fields[3].get_attribute('textContent')
            extracted_record['البريد الالكتروني'] = fields[4].get_attribute('textContent')

            # Append record to all details
            all_details.append(extracted_record)

            

    except Exception as e:
        # Close the browser if we run to an error
        print(e)
        driver.close()
        
    try:
        next_button = driver.find_element_by_css_selector("a[title='التالي']").get_attribute('href')
        driver.execute_script(next_button[11:-1])
        time.sleep(2)
        print('done')
            
    except Exception as e:
        # Close the browser if we run to an error
        print(e)
        break
```

```python
df = pd.DataFrame(all_details)
df
```

```python
# Save the information into a CSV file
df.to_csv("Factories_Table.csv", sep=',', encoding="utf-8-sig", index=False)
```
